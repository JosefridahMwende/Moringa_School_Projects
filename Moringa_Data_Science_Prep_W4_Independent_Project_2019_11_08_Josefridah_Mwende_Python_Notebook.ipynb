```
#importing the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import pandas_profiling as pp
import matplotlib.pyplot as plt
```

```
#reading and reviewing the dataset
Autolib = pd.read_csv('Autolib_dataset (2).csv', encoding='latin1')
Autolib.iloc[:5]
```

```
#getting to know more about our dataset
print(Autolib.info())
```

```
#standardization and fixing messy errors
Autolib.columns = Autolib.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
Autolib.iloc[:5]
```

```
#checking for null values and dropping unnecessary columns
Autolib.isna().sum()
```

```
# dropping unnecessary columns
Autolib.drop(['displayed_comment','scheduled_at',], axis=1, inplace=True)
Autolib.iloc[:5]
```

```
#checking for duplicate values and dropping duplicates
Autolib.duplicated()
Autolib.drop_duplicates(keep = False, inplace = True)
Autolib.shape
```

```
#confirming that there are no null values
np.count_nonzero(Autolib.isnull())
```

```
#understanding unique values of the dataset
print(Autolib['status'].unique().tolist())
print(Autolib['rental_status'].unique().tolist())
print(Autolib['subscription_status'].unique().tolist())
print(Autolib['kind'].unique().tolist())
print(Autolib['charging_status'].unique().tolist())
print(Autolib['station_type'].unique().tolist())
print(Autolib['status'].unique().tolist())
```

```
#making sure the city is Paris
Autolib_Paris = Autolib[(Autolib['city'] == "Paris")]
Autolib_Paris
```

```
#merging the year,month,day,,hour, minute columns into one
Autolib_Paris['Timestamp'] = pd.to_datetime(Autolib_Paris[['day','month','year','hour','minute']])
Autolib_Paris
```

```
#grouping using id and timestamp and hour
Autolib_Paris =  Autolib_Paris.sort_values(['Timestamp']).groupby(['id','Timestamp','hour']).sum()
Autolib_Paris
```

```
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) 
# in the city of Paris over the month of April 2018
Autolib_Paris['diff'] = Autolib_Paris.groupby(['id'])['bluecar_counter'].apply(lambda x:  x.shift(1) - x)
Autolib_Paris['diff']
Autolib_Paris.iloc[:10]
```

```
#negative diff means returning and positive difference means picking up
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018
Autolib3  = Autolib_Paris[(Autolib_Paris['diff'] > 0)].groupby(['hour'])['diff'].sum().sort_values(ascending = False)
Autolib3.iloc[:1]
```

```
# What is the most popular hour for returning cars?
Autolib4  = Autolib_Paris[(Autolib_Paris['diff'] < 0)].groupby(['hour'])['diff'].sum().sort_values(ascending = True)
Autolib4.iloc[:1]
```

```
#most popular station
Autolib_Paris['diff1'] = Autolib_Paris['diff'].abs()
Autolib_Paris.groupby(['id'])['diff1'].sum().sort_values(ascending = False).iloc[:1]
```
