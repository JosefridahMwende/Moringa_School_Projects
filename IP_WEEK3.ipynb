```python
#splitting column names
df2.columns.str.split(';').str[0].tolist()
['VILLES','STATUS','LOCALISATION','DECOUPZONE','ZONENAME','LONGITUDE','LATITUDE','REGION','AREA','CELL_ID','SITE_CODE']
```

```python
#merging 1, 2, 3 datasets
frames = [df3, df4, df5]
df6 = pd.concat(frames)
df6.head()
```

```python
#comparing the merged data sets to the geo data set

df7 = pd.merge(df6,df2, on =('CELL_ID','SITE_ID'), how = 'outer')
df7.head()
```

```python
#cleaning my data

df7.drop_duplicates() #checking for duplicate values
df7.isna() #checking for missing values number
df7.drop(['Unnamed: 0', 'VILLES','STATUS','LOCALISATION','DECOUPZONE','ZONENAME','LONGITUDE','LATITUDE','REGION','AREA'], axis = 1) #dropping columns
df8 = df7.dropna(how = "any") #drops records with any missing values
df8.head()
```
